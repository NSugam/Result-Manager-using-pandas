```python
import os
print (30* "*","Result Manager | Sugam Neupane | 1CR20CS196","*" *30)


def new():            # Storing data in csv file
    
    global data
    data = ""
    n = int(input("Number of Students: "))
    for i in range(n):
        name = input("Full Name of Student: ").title()
        phy = int(input("Mark obtained in Physics: "))
        chem = int(input("Mark obtained in Chemistry: "))
        math = int(input("Mark obtained in Maths: "))
        print (30* "*")
        total = phy+chem+math
        perc = format(total/3,".2f")
        info = f"{name},{phy},{chem},{math},{total},{perc}\n"
        data = data + info
    store()

def store():          # Asking file name from user and storing the data
    
    file_name = input("Enter the name of file: ")
    
    if os.path.exists(f"data/{file_name}.csv"):
        ask_overwrite = input("File already exist.\n Do you wish to overwrite(Y/N)? :").lower()
        if ask_overwrite == 'y':
            b = open(f"data/{file_name}.csv",'w')
            b.write('Name,Physics,Chemistry,Maths,Total,Percentage\n')
            b.write(data)
            b.close()
            print (f" Data Overwritten. Data Stored on: {file_name}.csv")
            print (30* "*")
            ask()
        elif ask_overwrite == 'n':
            store()
            
    else:
        b = open(f"data/{file_name}.csv",'x')
        b.write('Name,Physics,Chemistry,Maths,Total,Percentage\n')
        b.write(data)
        b.close()
        print (f"Data Store Successful. Data stored on: {file_name}.csv\n")
        print (30* "*")
        ask()

            
def read():
    file_name = input("Enter the name of file: ")
    if os.path.exists(f"data/{file_name}.csv"):
        print (30* "*",f"Record stored on: {file_name}.csv |","*" *30)
        import pandas as pd
        df = pd.read_csv(f'data/{file_name}.csv',index_col = 'Name') #usecols = ['Name','Maths']
        df.iloc[0::] #index localizer'
        print("\n",df,"\n")
        ask()
    else:
        print("\nError, File not found !!")
        ask()

    
def modify():
    file_name = input("Name of file to add data: ")
    try:
        b = open(f"data/{file_name}.csv")
        adata = ""
        n = int(input("Number of Students: "))
        for i in range(n):
            name = input("Full Name of Student: ").title()
            phy = int(input("Mark obtained in Physics: "))
            chem = int(input("Mark obtained in Chemistry: "))
            math = int(input("Mark obtained in Maths: "))
            print (30* "*")
            total = phy+chem+math
            perc = format(total/3,".2f")
            info = f"{name},{phy},{chem},{math},{total},{perc}\n"
            adata = adata + info
        
        b = open(f"data/{file_name}.csv",'a')
        b.write(adata)
        b.close()
        ask()
    except:
        print("\nError, File not found. Try creating a new file.\n")
        ask()


def delete():

    file_name = input("Name of file to delete: ")
    if os.path.exists(f"data/{file_name}.csv"):
        delt = input("Are you sure (y/n)?: ").lower()
        if delt == "y":
            os.remove(f"data/{file_name}.csv")
            print(f"\n{file_name}.csv has been deleted.\n")
            ask()
        elif delt == "n":
            ask()
        else:
            print("Input error !!")
            delete()
    else:
        print("File not found.")
        ask()


def end():
    print (30* "*","Thanks for using my system.","*" *30)

def ask():
    a = (input("1. Create new Data \n2. And or Modify Data \n3. Read Data  \n4. Delete data \n5. Terminate Program:\n"))   
    
    if a == "1":
        new()
    elif a == "2":
        modify()
    elif a == "3":
        read()
    elif a == "4":
        delete()
    elif a == "5":
        end()
    else:
        print ("Input error.")
        ask()
ask()
```
